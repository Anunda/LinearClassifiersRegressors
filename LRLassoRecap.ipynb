```python
from sklearn.linear_model import Lasso
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
random_state=24
boston = datasets.load_boston()
print(boston.keys())
boston.feature_names
X = boston.data
y = boston.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=random_state)
```

```python
lasso = Lasso()
lasso.fit(X_train,y_train)
y_preds = lasso.predict(X_test)
print('LR Score: ',lasso.score(X_test,y_test))
print('root mean squared error',np.sqrt(mean_squared_error(y_preds,y_test)))
```

SELECTING IMPORTANT FEATURES USING LASSO

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.set()
lasso_coef = lasso.fit(X_train,y_train).coef_
ax.plot(range(len(boston.feature_names)),lasso_coef)
plt.xticks(range(len(boston.feature_names)),boston.feature_names)
plt.show()
```
